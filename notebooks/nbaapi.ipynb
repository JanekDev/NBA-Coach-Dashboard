```python
from nba_api.stats.endpoints import playercareerstats, commonteamroster
from nba_api.stats.static import teams, players
import pandas as pd

# Get team and players data
nba_teams = teams.get_teams()
nba_players = players.get_players()

team_id = [team['id'] for team in nba_teams if team['full_name'] == 'Atlanta Hawks'][0]

# Fetch team roster
roster = commonteamroster.CommonTeamRoster(team_id=team_id, season='2022-23')
roster = roster.get_data_frames()[0]

# Select advanced stats of interest
advanced_stats = ['PLAYER_ID', 'LEAGUE_ID', 'Team_ID', 'GP', 'GS', 'MIN', 'FGM', 'FGA', 'FG_PCT', 'FG3M', 'FG3A', 'FG3_PCT', 'FTM', 'FTA', 'FT_PCT', 'OREB', 'DREB', 'REB', 'AST', 'STL', 'BLK', 'TOV', 'PF', 'PTS']

# For each player on the roster, calculate the average of advanced stats over last 3 seasons
players_stats = []

for index, player in roster.iterrows():
    player_id = player['PLAYER_ID']
    career = playercareerstats.PlayerCareerStats(player_id=player_id).get_data_frames()[0]
    
    # Select last 3 seasons (if available)
    career = career.tail(3)
    print(career.columns)
    # Calculate the mean of the stats
    mean_stats = career[advanced_stats].mean()
    mean_stats['PLAYER_NAME'] = player['PLAYER']
    players_stats.append(mean_stats)

players_stats_df = pd.DataFrame(players_stats)

# Define positions
positions = ['PG', 'SG', 'SF', 'PF', 'C']

# For each position, select the player with highest PER
starting_five = {}

for position in positions:
    players_in_position = players_stats_df[players_stats_df['POSITION'] == position]
    best_player = players_in_position.loc[players_in_position['PER'].idxmax()]['PLAYER_NAME']
    starting_five[position] = best_player

print(starting_five)
```
